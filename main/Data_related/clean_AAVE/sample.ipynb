```python
import pandas as pd
import re

# Paths
aa_file = r"C:\leo\college\RIT\my learning journey\hakyung sung\ASC-grammar-feedback\Bias\Stage1\Data\TwitterAAE-full-v1\twitteraae_all_aa"
output_file = r"C:\leo\college\RIT\my learning journey\hakyung sung\ASC-grammar-feedback\Bias\Stage1\Data\cleaned_data\aae_extracted_cleaned.csv"

# Column names in the file
columns_all = [
    "tweet_id", "timestamp", "user_id", "coordinates", "blockgroup",
    "text", "p_AA", "p_Hispanic", "p_Other", "p_White"
]

# Columns we want to keep
columns_keep = ["tweet_id", "text", "p_AA", "p_Hispanic", "p_Other", "p_White"]

# Cleaning function
def clean_tweet(text):
    text = str(text).lower()
    text = re.sub(r'@\w+', '', text)           # remove usernames
    text = re.sub(r'http\S+', '', text)        # remove URLs
    text = re.sub(r'#\w+', '', text)           # remove hashtags
    text = re.sub(r'&\w+;', '', text)          # remove HTML entities
    text = re.sub(r'[^a-z\s\'!?.,]', '', text) # keep letters and basic punctuation
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Load the AA subset robustly
df = pd.read_csv(
    aa_file,
    sep='\t',
    header=None,
    names=columns_all,
    quoting=3,            # ignore quote chars
    encoding='utf-8',     # handle special characters
    on_bad_lines='skip'   # skip lines with wrong number of columns
)

# Keep only required columns
df = df[columns_keep].copy()

# Clean tweet text
df['text'] = df['text'].apply(clean_tweet)

# Filter out tweets with fewer than 4 words
df = df[df['text'].str.split().str.len() >= 4]

# Save to CSV
df.to_csv(output_file, index=False)

print(f"Saved {len(df)} cleaned AA tweets to {output_file}")
```
